### Advance Patterns

- 1 Double Staircase with reversal

```python
n = 4

def advance_pattern(n):
    for i in range(1,n+1):
        
        # left side numbers
        for j in range(1,i+1):
            print(i,end='')
        
        # spaces formula
        for j in range((2*n)-(2*i)):
            print(' ',end='')
            
        # right side numbers
        for j in range(i,0,-1):
            print(j,end='')
        
        print()
                        
```

```python
advance_pattern(n)
```

- 2. Incremental Staircase

```python
def incremental_staircase(n):
    
    cnt =1

    # Outerloop
    for i in range(1,n+1):
        # for inner numbers logic
        for j in range(1,i+1):
            print(cnt, end='')
            cnt+=1
            
        print()
```

```python
incremental_staircase(5)
```

- 3. Alphabet Staircase
    - Note : It is same as numerical staricase but we need to use "char" keyword to get alphabets

```python
def alphabet_staircase(n):
    
    # Outerloop
    
    for i in range(n):
        
        # A char is 65 so we will start with 65 to line cnt numerical
        for j in range(65,66+i):
            print(chr(j),end=' ')
        
        print()
```

```python
alphabet_staircase(5)
```

- 3. Reverse Alphabet case 
    - Note : It is same as logic to get alphabets with decremental outerloop

```python
def reverse_alphabet_case(n):
    
    # outerloop
    for i in range(n+1,0,-1):
        
        # same logic wrt to previou pattern 
        for j in range(65,65+i):
            print(chr(j),end='')
        print()
```

```python
reverse_alphabet_case(n)
```

- 6 Invert Alphabetical pattern

```python
def invert_alphabetical_pattern(n):
    
    # outerloop
    for i in range(1,n+1):
        
        # reverse alphabetics
        for j in range((65+n-i),(65+n+1)):
            print(chr(j), end='')
        print()
            
```

```python
invert_alphabetical_pattern(5)
```

- 7. Diamond Pattern

```python
def diamond_pattern(n):
    
    # first half
    for i in range(n,0,-1):
        print(i*'*'+ (n-i)*' '+(n-i)*' '+i*'*')
    
    # second half
    for i in range(1,n+1):
        print(i*'*'+ (n-i)*' '+(n-i)*' '+i*'*')
```

```python
diamond_pattern(5)
```

- 9. square

```python
def square_pattern(n):
    
    for i in range(n+1):
        if i==0 or i ==n:
            print(n*'*')
        else:
            print('*'+(n-2)*' '+'*')
```

```python
n = 5
square_pattern(n)
```
